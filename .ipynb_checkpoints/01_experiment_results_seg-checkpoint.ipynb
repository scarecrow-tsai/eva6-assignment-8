#### Base Dataset Class

```python
class DatasetChestXrayMontShznJsrt(Dataset):
    def __init__(                                                                                                                                                            
        self,                                                                                                                                                                
        images_path,                                                                                                                                                         
        masks_path,                                                                                                                                                          
        image_transform,                                                                                                                                                     
        img_msk_size,                                                                                                                                                        
        dataset_name,                                                                                                                                                        
        dsubset="full",                                                                                                                                                      
    ):                                                                                                                                                                       
        super(DatasetChestXrayMontShznJsrt, self).__init__()                                                                                                                 
                                                                                                                                                                             
        self.images_path = pathlib.Path(images_path)                                                                                                                         
        self.masks_path = pathlib.Path(masks_path)                                                                                                                           
                                                                                                                                                                             
        self.dsubset = dsubset                                                                                                                                               
                                                                                                                                                                             
        self.img_msk_size = img_msk_size                                                                                                                                     
                                                                                                                                                                             
        self.transform = image_transform                                                                                                                                     
                                                                                                                                                                             
        if dataset_name == "shenzhen" or dataset_name == "montgomery":                                                                                                       
            if self.dsubset == "full":                                                                                                                                       
                self.img_file_paths = [file for file in images_path.glob("*.png")]                                                                                           
                self.msk_file_paths = [file for file in masks_path.glob("*.png")]
            elif self.dsubset == "normal":
                self.img_file_paths = [file for file in images_path.glob("*0.png")]
                self.msk_file_paths = [file for file in masks_path.glob("*0.png")]
            elif self.dsubset == "abnormal":
                self.img_file_paths = [file for file in images_path.glob("*1.png")]
                self.msk_file_paths = [file for file in masks_path.glob("*1.png")]
            else:
                raise ValueError("Invalid dsubset value.")
        elif dataset_name == "jsrt":
            if self.dsubset == "full":
                self.img_file_paths = [file for file in images_path.glob("*.png")]
                self.msk_file_paths = [file for file in masks_path.glob("*.png")]
            elif self.dsubset == "normal":
                self.img_file_paths = [file for file in images_path.glob("*.png")]
                self.img_file_paths = [
                    file for file in self.img_file_paths if file.stem[3] == "N"
                ]
                self.msk_file_paths = [file for file in masks_path.glob("*.png")]
                self.msk_file_paths = [
                    file for file in self.msk_file_paths if file.stem[3] == "N"
                ]
            elif self.dsubset == "abnormal":
                self.img_file_paths = [file for file in images_path.glob("*.png")]
                self.img_file_paths = [
                    file for file in self.img_file_paths if file.stem[3] == "L"
                ]
                self.msk_file_paths = [file for file in masks_path.glob("*.png")]
                self.msk_file_paths = [
                    file for file in self.msk_file_paths if file.stem[3] == "L"
                ]
            else:
                raise ValueError("Invalid dsubset value.")
        else:
            raise ValueError("Invalid dataset_name")
            
    def __getitem__(self, idx):
        img_path = str(self.img_file_paths[idx])
        msk_path = str(self.msk_file_paths[idx])
        img = (
            Image.open(img_path)
            .convert("L")
            .resize((self.img_msk_size, self.img_msk_size), Image.BICUBIC)
        )
        img = self.transform(img)
        msk = (
            Image.open(msk_path)
            .convert("L")
            .resize((self.img_msk_size, self.img_msk_size), Image.NEAREST)
        )
        msk = np.array(msk)
        msk[msk == 255] = 1
        msk = torch.from_numpy(msk).long()
        return img, msk

    def __len__(self):
        assert len(self.img_file_paths) == len(self.msk_file_paths)
        return len(self.img_file_paths)
```

## Test Models

```python
def make_predictions(data_loader, model):
    y_pred_list = []
    y_true_list = []
    x_true_list = []
    
    with torch.no_grad():
        for x_batch, y_batch in data_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            
            y_test_pred = model(x_batch)

            y_test_pred = torch.log_softmax(y_test_pred, dim=1)
            _, y_pred_tag = torch.max(y_test_pred, dim = 1)

            y_pred_list.append(y_pred_tag.cpu().numpy())
            y_true_list.append(y_batch.cpu().numpy())
            x_true_list.append(x_batch.squeeze().cpu().numpy())
            
    return x_true_list, y_true_list, y_pred_list


def calculate_iou(pred_list, true_list):
    total_iou = 0

    for i in range(len(pred_list)):
        temp_iou = jaccard_score(true_list[i].squeeze().flatten(), pred_list[i].squeeze().flatten(), average="macro")
        total_iou += temp_iou
        
        
    return total_iou/len(pred_list)



def array_to_img_for_viz(arr):
    arr = arr * 100
    arr = arr.squeeze()
    arr = np.uint8(arr)
    arr_img = Image.fromarray(arr)
    
    return arr_img  



def plot_img(pred_mask, true_mask, img_idx):

    image_index_to_view = img_idx

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))
    axes[0].imshow(array_to_img_for_viz(true_mask[image_index_to_view]))
    axes[0].set_title("True Mask")
    axes[1].imshow(array_to_img_for_viz(pred_mask[image_index_to_view]))
    axes[1].set_title("Predicted Mask")
    
    
    
def plot_img_grid(true_image_list, true_mask_list, pred_mask_list, img_idx_list, source_data, target_data):
    n = len(img_idx_list)
    
    fig, axes = plt.subplots(nrows=n, ncols=3, figsize=(15, n*3))
    for i, idx in enumerate(img_idx_list):
        axes[i, 0].imshow(array_to_img_for_viz(true_image_list[idx]))
        axes[i, 0].set_title(f"True Image | {target_data} #{idx}")
        axes[i, 1].imshow(array_to_img_for_viz(true_mask_list[idx]))
        axes[i, 1].set_title(f"True Mask | {target_data} #{idx}")
        axes[i, 2].imshow(array_to_img_for_viz(pred_mask_list[idx]))
        axes[i, 2].set_title(f"Predicted Mask | {target_data} #{idx}")
        
    fig.subplots_adjust(left=0.15, top=0.92, bottom=0.05)
    plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[])
    
    plt.suptitle(f"Source: {source_data} | Target: {target_data}", fontsize=14)
    
    
    
def run_test_suite(target_data_loader, source_model):
    true_image, true_mask, pred_mask  = make_predictions(target_data_loader, source_model)
    iou = calculate_iou(pred_mask, true_mask)
    
    return true_image, true_mask, pred_mask, iou
```

```python
test_models = {
    "model_mont_full": model_mont_full,
#     "model_shzn_full": model_shzn_full, 
#     "model_jsrt_full": model_jsrt_full,
}


test_dataloaders = {
    "data_mont_full": mont_loader_full, 
    "data_shzn_full": shzn_loader_full, 
    "data_jsrt_full": jsrt_loader_full,
}

results_df = pd.DataFrame(columns=["source", "target", "model", "iou"])
```

```python
for src_mdl in tqdm(test_models):
    for tgt_dat in tqdm(test_dataloaders):
        _, _, _, iou = run_test_suite(source_model=test_models[src_mdl], 
                                      target_data_loader=test_dataloaders[tgt_dat]
                                     )
        
        print(src_mdl,"  -  ", tgt_dat, "  -  ", iou)
        row_dict = {"source": src_mdl.split("_")[1] + "_" + src_mdl.split("_")[2],
                    "target": tgt_dat.split("_")[1] + "_" + tgt_dat.split("_")[2],
                    "model": MODEL_NAME,
                    "iou": iou
                   }
        
        results_df = results_df.append(row_dict, ignore_index=True)
```
